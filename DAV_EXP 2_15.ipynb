# Aim: Simple Linear Regression


# **PYTHON**

```r
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import datasets

# Load the diabetes dataset from scikit-learn
diabetes = datasets.load_diabetes()
data = pd.DataFrame(data=np.c_[diabetes['data'], diabetes['target']], columns=diabetes['feature_names'] + ['target'])

# Display the summary of the dataset
print("Summary of the Diabetes Dataset:")
print(data.head())
```

```r
# Check normality of the dependent variable (target)
plt.figure(figsize=(8, 5))
plt.hist(data['target'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of the Dependent Variable (Target)')
plt.xlabel('Target')
plt.ylabel('Frequency')
plt.show()
```

```r
# Check linearity between independent and dependent variable
plt.figure(figsize=(8, 5))
plt.scatter(data['bmi'], data['target'], color='orange', edgecolor='black')
plt.title('Relationship between BMI and Target')
plt.xlabel('BMI')
plt.ylabel('Target')
plt.show()
```

```r
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data['bmi'].values.reshape(-1, 1), data['target'].values, test_size=0.2, random_state=42)

# Perform Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)
```

```r
# Visualize the results
plt.figure(figsize=(8, 5))
plt.scatter(X_test, y_test, color='orange', edgecolor='black', label='Actual Data')
plt.plot(X_test, model.predict(X_test), color='blue', linewidth=2, label='Linear Regression')
plt.title('Linear Regression Analysis on BMI and Target')
plt.xlabel('BMI')
plt.ylabel('Target')
plt.legend()
plt.show()
```
